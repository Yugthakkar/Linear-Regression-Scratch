```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('data.csv')

def loss_function(m,b,points):
    total_error=0
    for i in range(len(points)):
        x=points.iloc[i].x
        y=points.iloc[i].y
        total_error += (y-(m*x+b))**2 # sum of squared errors
        total_error = total_error/float(len(points))# average of squared errors

def gradient_descent(m1,b1,points,L):
    m_grad=0
    b_grad=0

    n=len(points)
    for i in range(n):
        x=points.iloc[i].x
        y=points.iloc[i].y
        m_grad += -(2/n)*x*(y-(m1*x+b1))# partial derivative of m
        b_grad += -(2/n)*(y-(m1*x+b1))# partial derivative of b
    m=m1-(L*m_grad)
    b=b1-(L*b_grad)
    return m,b
m=0
b=0
l=0.0001
epochs=500
for i in range(epochs):
    if i%50==0:
        print(f"Epoch:{i}")
    m,b=gradient_descent(m,b,data,l)

print(m,b)
plt.scatter(data["x"], data["y"], color="purple", label="Data Points",marker="+")
plt.plot(data["x"], [m * x + b for x in data["x"]], color="blue", label="Best Fit Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```
